### Data Preprocessing

##### One Hot Encode all Categorical variables

```python
# # This one hot encoding is done outside the function because of the need to add a prefix
# train = pd.concat([train, pd.get_dummies(train.No_of_previous_employers, prefix="No_of_previous_employers")], axis=1)
# test = pd.concat([test, pd.get_dummies(test.No_of_previous_employers, prefix="No_of_previous_employers")], axis=1)
# train.drop(['No_of_previous_employers'], axis=1, inplace=True)
# test.drop(['No_of_previous_employers'], axis=1, inplace=True)
```

```python
# Convert all Categorical Data into Numerical Data
def convert_feature(data):
    if data.dtype == object:
        dummy_df = pd.get_dummies(data, prefix=data.name)
        new_train = pd.concat([train, dummy_df], axis=1)
        new_train.drop([data.name], axis=1, inplace=True)
    else:
        pass
    return new_train
    
```

```python
# Start the Encoding
train = convert_feature(train.Division)
train = convert_feature(train.Qualification)
train = convert_feature(train.Gender)
train = convert_feature(train.Channel_of_Recruitment)
train = convert_feature(train.State_Of_Origin)
train = convert_feature(train.Foreign_schooled)
train = convert_feature(train.Marital_Status)
train = convert_feature(train.Past_Disciplinary_Action)
train = convert_feature(train.Previous_IntraDepartmental_Movement)
train = convert_feature(train.No_of_previous_employers)
train['promoted_or_not'] = train.Promoted_or_Not
train.drop(['Promoted_or_Not'], axis=1, inplace=True)
```

```python
# Start Encoding Test data
def convert_test_feature(data):
    if data.dtype == object:
        dummy_df = pd.get_dummies(data, prefix=data.name)
        new_test = pd.concat([test, dummy_df], axis=1)
        new_test.drop([data.name], axis=1, inplace=True)
    else:
        pass
    return new_test



test = convert_test_feature(test.Division)
test = convert_test_feature(test.Qualification)
test = convert_test_feature(test.Gender)
test = convert_test_feature(test.Channel_of_Recruitment)
test = convert_test_feature(test.State_Of_Origin)
test = convert_test_feature(test.Foreign_schooled)
test = convert_test_feature(test.Marital_Status)
test = convert_test_feature(test.Past_Disciplinary_Action)
test = convert_test_feature(test.Previous_IntraDepartmental_Movement)
test = convert_test_feature(test.No_of_previous_employers)
```

### Start Modeling
Train with all features first to check the accuracy


```python
# Create the feature and the target
X = train.iloc[:, 1:-1]
y = train.promoted_or_not
```

```python
print(X.shape)
print(y.shape)
```

##### We will be using cross val score to check performance instead of train test split

```python
# Import all models required to train and instantiate all
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

logreg_model = LogisticRegression(solver='liblinear')
decision_model = DecisionTreeClassifier()
knn_model = KNeighborsClassifier(n_neighbors=11)
random_model = RandomForestClassifier(criterion='entropy', n_estimators=100)
```

```python
from sklearn.model_selection import cross_val_score

models = [logreg_model, decision_model, knn_model, random_model ]
for model in models:
    scores = cross_val_score(model, X, y, cv=10, scoring='accuracy')
    print("Accuracy for ", model,  " is ", scores.mean())
```

```python
random_model.fit(X, y)
```

```python
test = test.iloc[:, 1:]
prediction = random_model.predict(test)
```
